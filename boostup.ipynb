# Is there a correlation between years of experience and salary?

```python
# Scatter Plot
x=merged_data['YearsExperience'].values
y=merged_data['Salary'].values
plt.scatter(x, y)
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title('Correlation between years of experience and Salary')
plt.show()
```

```python
x_mean=np.mean(x)
y_mean=np.mean(y)
```

```python
## calculation of line's m and b 
x_num = x-x_mean
y_num = y-y_mean
m=(sum(x_num*y_num))/(sum(x_num*x_num))
```

```python
m
```

```python
b=y_mean-(m*x_mean)
b
```

```python
# Drawing the Linear Regression Line
y1 = m*x+b
plt.plot(x, y1, '-r',label='Best Fit Line')
plt.scatter(x, y)
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title('Correlation between years of experience and Salary')
plt.show()
```
